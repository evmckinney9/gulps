```python
from io import BytesIO
import matplotlib.pyplot as plt
from hetero_isas.monodromy_lp import (
    MonodromyLPDecomposer,
)
from hetero_isas.monodromy_lp.isa import ISAHandler
from PIL import Image
from qiskit.circuit.library import CXGate, CZGate, SwapGate, iSwapGate
from qiskit.quantum_info.random import random_unitary
from tqdm.notebook import tqdm
```

```python
isa_handler = ISAHandler(
    [
        CXGate(),
        iSwapGate(),
        CXGate().power(1 / 2),
        iSwapGate().power(1 / 2),
        CXGate().power(1 / 3),
        iSwapGate().power(1 / 3),
        # SwapGate(),
    ],
    [1.0, 2, 1 / 2, 1, 1 / 3, 2 / 3],
    ["cx", "iswap", "sq[2]cx", "sq[2]iswap", "sq[3]cx", "sq[3]iswap"],
    # ["iswap", "sq[2]iswap", "sq[3]iswap", "swap"],
)
decomposer = MonodromyLPDecomposer(isa_handler)
```

```python
imlist = []
for _ in tqdm(range(100)):
    target = random_unitary(4).to_matrix()
    result = decomposer._best_decomposition(target)
    ax = result.render_path()
    buf = BytesIO()
    plt.savefig(buf, format="png")
    buf.seek(0)
    im = Image.open(buf)
    imlist.append(im)
    plt.close()  # stops from rendering

# put everything together and save as a gif
imlist[0].save(
    "../../../docs/images/animated4.gif",
    save_all=True,
    append_images=imlist[1:],
    optimize=True,
    duration=250,
    loop=0,
)
```
