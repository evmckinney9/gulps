```python
%matplotlib inline
from hetero_isas.zz_parallel_drive.ansatz import (
    BasicCircuitAnsatz,
    PulseConstraintPolytopes,
)
from qiskit.circuit.library import iSwapGate
from hetero_isas.zz_parallel_drive.bgate import BGate
```

```python
ansatz = BasicCircuitAnsatz(BGate(), 3)
print(ansatz._to_string("B"))
ansatz.qc.draw("mpl");
```

```python
ansatz = BasicCircuitAnsatz(iSwapGate().power(1 / 2), 2, u_sequence=[(1, 1)])
ansatz.generate_points(10_000, "qucontrol")
PulseConstraintPolytopes._plot_weyl_polytope_scatter(ansatz=ansatz)
print(ansatz._to_string("sqiS"))
ansatz.qc.draw("mpl");
```
