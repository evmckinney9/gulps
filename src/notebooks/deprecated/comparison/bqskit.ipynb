```python
from tqdm import tqdm

# Gate set with multiple entangling gates
from bqskit.ir.gates import SqrtISwapGate, SycamoreGate, ISwapGate, CXGate, SqrtCNOTGate
from bqskit.ir.gates import CZGate, RZGate, SXGate, U3Gate
from bqskit import MachineModel

# Load it into BQSKit
from bqskit.ext import qiskit_to_bqskit

# Compile the circuit
from bqskit import compile

from bqskit.qis import UnitaryMatrix
```

```python
# Build a circuit in Qiskit
from qiskit.circuit.random import random_circuit

input_qc = random_circuit(2, 1, max_operands=2, num_operand_distribution={1: 0, 2: 1.0})


bqskit_circuit = qiskit_to_bqskit(input_qc)
```

```python
out_circuit = compile(bqskit_circuit)

# Print new statistics
print("Compiled Circuit Statistics")
print("Gate Counts:", out_circuit.gate_counts)
print("Logical Connectivity:", out_circuit.coupling_graph)
```

```python
# Define a function to save space
def compile_to_gateset(circuit, gate_set):
    model = MachineModel(2, gate_set=gate_set)
    return compile(circuit, model=model)


gate_set = {SqrtISwapGate(), ISwapGate(), CXGate(), SqrtCNOTGate(), U3Gate()}

out_circuit = compile_to_gateset(bqskit_circuit, gate_set)
print(out_circuit.gate_counts)
```
